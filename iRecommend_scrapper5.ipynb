Парсинг отзывов с сайта iRecommend

Перед запуском кода надо вставить ссылки на страницы с отзывами в список urls (он в самом начале кода, сразу после импортов, сейчас в нем ссылки на отзывы про ношпу). Рекомендуется добавлять не более 3х ссылок, иначе сервер может отключить подключение. Данным кодом можно пользоваться примерно 1 раз в сутки. Если пользоваться чаще, сервер будет отключать подключение.

```python
import pandas as pd
```

```python
import requests as rq
from bs4 import BeautifulSoup
import time

urls = ['https://irecommend.ru/content/krem-dlya-litsa-nochnoi-vichy-liftactiv-collagen-specialist-night',
        'https://irecommend.ru/content/krem-dlya-litsa-nochnoi-vichy-liftactiv-collagen-specialist-night?page=1',
        'https://irecommend.ru/content/krem-dlya-litsa-nochnoi-vichy-liftactiv-collagen-specialist-night?page=2',
        'https://irecommend.ru/content/krem-dlya-litsa-nochnoi-vichy-liftactiv-collagen-specialist-night?page=3',
        'https://irecommend.ru/content/krem-dlya-kozhi-vokrug-glaz-vichy-liftactiv-collagen-specialist-protiv-5-priznakov-stareniya'

]

# Define headers with a User-Agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
all_links = []
for u in urls:
    page = rq.get(u, headers=headers)
    time.sleep(2)  # Consider reducing this if not necessary

    if page.status_code != 200:
        time.sleep(3)
        print(f"Failed to retrieve {u} - Status Code: {page.status_code}")
        continue
        time.sleep(1)

    text = page.text.split('Смотрите также')
    time.sleep(2)
    soup = BeautifulSoup(text[0], 'html.parser')
    time.sleep(1)

    l = []
    for i in soup.find_all('a'):
        time.sleep(2)
        if i.get('data-text') == 'Читать весь отзыв':
            time.sleep(4)
            href = i.get('href')
            time.sleep(1)
            if href:  # Check if href is not None
                time.sleep(2)
                l.append(href)
                time.sleep(1)

    main = 'https://irecommend.ru'
    full_link = [main + link for link in l]  # List comprehension for full links
    time.sleep(2)

    all_links.extend(full_link)  # Flatten the list
    time.sleep(1)

print("All links found:", all_links)
```

```python
len(all_links)
```

```python
def GetTexts(url0):
  page0 = rq.get(url0)
  time.sleep(2)
  soup0 = BeautifulSoup(page0.text, features="html.parser")
  time.sleep(1)
  text_list = soup0.find_all('p')
  time.sleep(3)
  text = [i.text for i in text_list]
  time.sleep(1)
  final_text = ' '.join(text)
  time.sleep(2)
  final_text = final_text.replace('\xa0', ' ')
  time.sleep(1)
  return url0, final_text
```

```python
texts = []

for link in all_links:
  text = GetTexts(link)
  texts.append(text)
```

```python
df = pd.DataFrame(texts)
df.head()
```

```python
df.columns = ['link', 'text']
```

```python
df = df.drop('link', axis=1)
```

```python
df.to_csv('Liftactiv_Collagen_specialist_corpus.csv')
```
